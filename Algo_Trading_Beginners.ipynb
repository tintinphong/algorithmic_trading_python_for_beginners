## Creating a Strategy (SMA)

```python
apple.columns = ["Close", "d_returns", "cummreturns", "cummmax", "drawdown", "drawdown%"]
#rename AAPL to Close
apple
```

```python
data = apple.Close.loc[(apple.index>="1991-01-01")]
data
```

```python
data = data.to_frame()
```

```python
data
```

### Golden Cross strategy:
-When short term moving average is higher than long term moving average it signals potential shift so we enter long positions

-When short term moving average is lower than long term moving average we exit


```python
sma_s = 50
sma_l = 100
```

```python
data["sma_s"] = data.Close.rolling(sma_s).mean()
data["sma_l"] = data.Close.rolling(sma_l).mean()
```

```python
data
```

```python
data.plot(figsize = (12,8), title = "AAPL - SMA{} | SMA{}".format(sma_s,sma_l), fontsize = 12)
```

```python
data.loc["2016"].plot(figsize = (12,8), title = "AAPL - SMA{} | SMA{}".format(sma_s,sma_l), fontsize = 12)
```

```python
data.dropna(inplace = True)
```

```python
data
```

```python
data["position"] = np.where(data["sma_s"]>data["sma_l"],1,-1)
```

```python
data
```

```python
data.loc["2016",["sma_s","sma_l","position"]].plot(figsize = (12,8), title = "AAPL - SMA{} | SMA{}".format(sma_s,sma_l), fontsize = 12,secondary_y = "position")
```

## Backtesting the SMA strategy

```python
data["returnsb&h"] = np.log(data.Close.div(data.Close.shift(1)))
#This is the data for the basic buy and hold strategy
```

```python
data["strategy"] = data["returnsb&h"]*data.position.shift(1)
#This is the return for the SMA strategy (Golden cross
#we have to shift one becase were taking signal from yesterday to apply it today
```

```python
data
```

```python
data.dropna(inplace = True)
```

```python
data
```

```python
data[["returnsb&h","strategy"]].sum()
```

```python
data[["returnsb&h","strategy"]].sum().apply(np.exp) #what 1 dollar investment will be
```

```python
data[["returnsb&h","strategy"]].std()*np.sqrt(252) #annual std
```

```python
data.Close.plot(figsize = (12,8))
```

```python
#general closing graph shows bias towards up size, we should create a strategy so we dont go short
#not exposing ourselfs to the choppiness of the short
```

### Strategy Adjusted with Long Bias

```python
data["position2"] = np.where(data["sma_s"]>data["sma_l"],1,0)
```

```python
data["strategy2"] = data["returnsb&h"]*data.position2.shift(1)
```

```python
data.dropna(inplace = True)
```

```python
data
```

```python
data[["returnsb&h","strategy2"]].sum()
```

```python
data[["returnsb&h","strategy2"]].sum().apply(np.exp) #what 1 dollar investment will be
#the reason the returnsb&h changes was from the dropna getting rid of the first 100 days return but it is okay
```

```python
data[["returnsb&h","strategy2"]].std()*np.sqrt(252) #annual std
```

```python
## Strategy 2 is better than Strategy 1 as it has higher returns and lower volatility
data[["strategy","strategy2"]].sum().apply(np.exp)
```
