```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tickers = ["PLTR", "NVDA", "SMCI", "MSTR", "CVNA"]
data = yf.download(tickers, period="max", auto_adjust=False)["Close"]
for ticker in tickers:
    plt.figure(figsize = (12,8)) #creates blank canvas for graph every loop
    stock_data = data[ticker].dropna() #removes NAs for ticker
    stock_data.plot(label = "Price", color = "black", alpha = 0.5) #label is to create a legend automatically
    stock_data.rolling(50).mean().plot(label = "SMA 50", color = "blue")
    stock_data.rolling(200).mean().plot(label = "SMA 200", color = "red")

    plt.title(f"{ticker} - Closing Price vs SMAs (Full History)")
    plt.legend(fontsize = 12)
    plt.grid(True, alpha=0.3)
    plt.show()
```
